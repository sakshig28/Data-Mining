### Import libraries

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

### Load data

```python
iris = load_iris()
```

### Explore data

```python
iris.keys()
```

```python
print(iris.DESCR)
```

```python
iris.feature_names
```

```python
iris.target_names
```

### Prepare data

```python
X = pd.DataFrame(iris.data)
y = pd.DataFrame(iris.target)
```

### Train model

```python
# Fit a DecisionTreeClassifier model
tree = DecisionTreeClassifier(max_depth = 2)     # Note max_depth value
tree.fit(X,y)
```

### Visualize model output

```python
# Visualize decision tree. Create dot file.
export_graphviz(
            tree,
            out_file =  'images/DT_1.dot',
            feature_names = iris.feature_names,
            class_names = iris.target_names,
            filled = True,
            rounded = True)
```

```python
# Linux command to draw a directed graph from a representation of decision tree (dot file).
! dot -Tpng images/DT_1.dot -o images/DT_1.png
```

```python
# Display directed graph
img = cv2.imread('images/DT_1.png')
plt.figure(figsize = (10, 10))
plt.imshow(img)
```

```python
# Make predictions on new sample 1
sample_one_pred = int(tree.predict([[5, 5, 1, 3]]))
print(f"The first sample most likely belongs a {iris.target_names[sample_one_pred]} flower.")
```

```python
# Make predictions on new sample 2
sample_two_pred = int(tree.predict([[5, 5, 2.6, 1.5]]))
print(f"The second sample most likely belongs a {iris.target_names[sample_two_pred]} flower.")
```

### Train model with different hyperparameters

```python
# Fit a DecisionTreeClassifier model with a different max_depth value
tree = DecisionTreeClassifier(max_depth = 5)
tree.fit(X,y)
```

### Visualize model output
